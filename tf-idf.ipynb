```python
def clean_en(li):
    clean_li = []
    for item in li:
        clean_li.append(item[:-2])
    return clean_li

def clean_ch(li):
    return li.remove("论文关键词")

def li_to_str(li):
    str = ""
    for item in li:
        str += item
        str += " "
    return str
```

```python
df_en = open('./key_words/output_en_keybert.txt',"r")
en_list = []
en_index = []
pattern_en = re.compile("[0-9a-zA-Z ]+[(]+")

index = 0
for line in df_en:
    en_words = clean_en(pattern_en.findall(line))
    if en_words == []:
        en_index.append(index)
        continue
    en_list.append(li_to_str(en_words))
    index += 1
```

```python
df_ch = open('./key_words/output_ch_keybert.txt',"r")
ch_list = []
ch_index = []
pattern_ch = re.compile("[\u4e00-\u9fa5]+")

index = 0
for line in df_ch:
    ch_words = pattern_ch.findall(line)
    if ch_words[0] == "学者":
        ch_index.append(index)
        continue
    ch_words.remove("论文关键词")
    ch_list.extend(ch_words)
    index += 1
```
