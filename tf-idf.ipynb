```python
from sklearn.feature_extraction.text import TfidfVectorizer
import os
import pandas as pd
import numpy as np
import re
```

```python
def clean_en(li):
    clean_li = []
    k=0
    for item in li:
        if k == 0:
            multiword = item[:-2].split(" ")
            clean_li.append(combine_word(multiword))
            k += 1
        else:
            multiword = item[1:-2].split(" ")
            clean_li.append(combine_word(multiword))

    
    return clean_li

def clean_ch(li):
    return li.remove("论文关键词")

def li_to_str_en(li):
    str = ""
    for item in li:
        str +=f"'{item}'"
        str += " "
    return str

def li_to_str_ch(li):
    str = ""
    for item in li:
        str += item
        str += " "
    return str

def combine_word(li):
    res = ''
    n = len(li)
    index = 0
    for item in li:
        index += 1
        res += item
        if index != n:
            res += '_'
    return res

def get_data_list(li,n):
    data_list = []
    tv = TfidfVectorizer(use_idf=True,smooth_idf=True)
    tv_fit = tv.fit_transform(li)
    words = tv.get_feature_names_out()
    weight=tv_fit.toarray()     
    for wei in weight:
        data_list.append(sorted(dict(zip(words,wei)).items(),key=lambda x:x[1],reverse=True)[:n])
    return data_list

def save_data(path,index_list,author_list,res):
    fp = open(path,'w',encoding='utf-8')
    n = len(author_list)
    cur = 0
    index = 0
    for author in author_list:
        fp.write(author_list[cur])
        fp.write('\n')
        if cur != n-1:
            line_num = index_list[cur+1]-index_list[cur]
        else:
            line_num = len(res) - index_list[cur]
        for i in range(line_num):
            fp.write("论文关键词")
            fp.write(str(res[index])[1:-1])
            fp.write('\n')
            index += 1
        cur += 1
    fp.close()
```

```python
en_path = './key_words/output_en_keybert.txt'
df_en = open(en_path,"r")
en_list = []
en_index = []
en_author = []
pattern_en = re.compile("[0-9a-zA-Z ]+[(]+")

index = 0
for line in df_en:
    en_words = clean_en(pattern_en.findall(line))
    if en_words == []:
        en_index.append(index)
        en_author.append(line[:-1])
        continue
    en_list.append(li_to_str_en(en_words))
    index += 1
```

```python
ch_path = './key_words/output_ch_keybert.txt'
df_ch = open(ch_path,"r")
ch_list = []
ch_index = []
ch_author = []
pattern_ch = re.compile("[\u4e00-\u9fa5]+")

index = 0
for line in df_ch:
    ch_words = pattern_ch.findall(line)
    if ch_words[0] == "学者":
        ch_index.append(index)
        ch_author.append(line[:-1])
        continue
    ch_words.remove("论文关键词")
    index += 1
    ch_list.append(li_to_str_ch(ch_words))

```

```python
#传入数据list和像提取的topN数量

ch_res = get_data_list(ch_list,10)
en_res = get_data_list(en_list,10)
```

```python
save_data('./key_words/ch_tfidf.txt',ch_index,ch_author,ch_res)
save_data('./key_words/en_tfidf.txt',en_index,en_author,en_res)
```
